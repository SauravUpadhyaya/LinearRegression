```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Define the range of k values to explore
k_range_iris = range(2, 7)

# Compute the sum of squared distances for each value of k in the range
sum_of_squared_distances_iris = []
for k in k_range_iris:
    km = KMeans(n_clusters=k)
    km.fit(x)
    sum_of_squared_distances_iris.append(km.inertia_)

# Plot the Elbow method for the Iris dataset
fig, ax = plt.subplots()
ax.plot(k_range_iris, sum_of_squared_distances_iris, 'gx-', linewidth=2)
ax.set_xlabel('Number of Clusters (k)', fontsize=14)
ax.set_ylabel('Sum of Squared Distances', fontsize=14)
ax.set_title('Elbow Method for Iris Dataset', fontsize=16)
ax.grid(True)
plt.show()
```
